```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
iris_df=sns.load_dataset('iris')
```

```python
sns.FacetGrid(iris_df,hue="species",height=3).map(sns.displot,"petal_length").add_lengend()
sns.FacetGrid(iris_df,hue="species",height=3).map(sns.displot,"petal_width").add_lengend()
sns.FacetGrid(iris_df,hue="species",height=3).map(sns.displot,"sepal_length").add_lengend()
sns.FacetGrid(iris_df,hue="species",height=3).map(sns.displot,"sepal_length").add_lengend()
plt.show()


```

```python
iris=load_iris()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(iris.data,iris.target,test_size=0.2,random_state=42)
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(X_train,Y_train)
```

```python
Y_pred=clf.predict(X_test)
```

```python
accuracy=clf.score(X_test,Y_test)
```

```python
plt.figure(figsize=(12,8))
```

```python
plot_tree(clf,filled=True,feature_names=iris.feature_names,class_names=iris.target_names)
plt.show()
```
